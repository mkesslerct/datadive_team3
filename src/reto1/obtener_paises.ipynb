```python
# Read file
profundidad_agua_respecto_superficie = pd.read_csv(
    DATA_DIR / "article_data" / "AnnualDepthToGroundwater.csv",
    sep=",",
    index_col=0
)

profundidad_agua_respecto_superficie
```

```python
# Add a column with the country name, using geopy to get the country from the coordinates
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="test")

def get_country_name(lat, lon):
    location = geolocator.reverse(f"{lat}, {lon}")
    return location.raw["address"]["country"]

profundidad_agua_respecto_superficie.apply(
    lambda row: get_country_name(row["Lat"], row["Lon"]), axis=1
)
```
