```python
import matplotlib.pyplot as plt

from sklearn.linear_model import TheilSenRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
# Select the relevant columns from the dataframe
data = el_pozo.dropna(subset=['Id', 'Punto', 'Caudal', 'Cota', 'Prof'])
y = data['Prof'].values
# X = data[['Id', 'Punto', 'Caudal', 'Cota']].values.reshape(-1, 4)
X = data[['PointX', 'PointY', 'Caudal', 'Cota', 'Prof']].values.reshape(-1, 5)

# Create and fit the Theil-Sen regression model
model = TheilSenRegressor()
model.fit(X, y)

# Predict the target values
y_pred = model.predict(X)

# Plot the actual and predicted values
plt.scatter(y, y_pred)
plt.xlabel('Actual Prof')
plt.ylabel('Predicted Prof')
plt.title('Theil-Sen Regression')
plt.show()
# Import necessary libraries
import matplotlib.pyplot as plt

# Select the relevant columns from the dataframe
data = el_pozo[['PointX', 'PointY', 'Caudal', 'Cota', 'Prof']].copy()

# Remove rows with missing values
```
